### Step 1: Load and Inspect the dataset 

```python
# Necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# load dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Convert into DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
X
```

```python
y
```

### Step 2 : Data preprocessing

```python
from sklearn.preprocessing import StandardScaler, LabelBinarizer
```

```python
scale = StandardScaler()
```

```python
X_scale = scale.fit_transform(X)
X_scale
```

```python
lb = LabelBinarizer()
```

```python
y_encoded = lb.fit_transform(y)
y_encoded
```

### Step 3 : Split the dataset

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scale, y_encoded, test_size = 0.2, random_state = 10 )
```

```python
X_train
```

```python
y_train
```

### Step 4 : Build the model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential([
    Dense(4, input_dim = 4, activation = 'sigmoid'),
    Dense(5, activation = 'relu'),
    # Dropout(0.2),
    # Dense(6, activation = 'relu'),
    # Dropout(0.3),
    Dense(3, activation = 'softmax')
])
```

### Step 5 : Compile the model

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

### Step 6 : Train the model

```python
early_stopping = EarlyStopping(monitor = 'val_loss', patience = 10)
```

```python
history = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 100, batch_size = 16, callbacks = [early_stopping], verbose = 1)
```

### Step 7 : Save the Model

```python
model.save("ANN_Model_iris.h5")
```

### Step 8 : Evaluate the model

```python
train_acc, train_loss = model.evaluate(X_train, y_train)
```

```python
test_acc, test_loss = model.evaluate(X_test, y_test)
```

### Step 9 : Accuracy Plot

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy'], label = 'train_accuracy')
plt.plot(history.history['val_accuracy'], label = 'test_accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Model Accuracy')
plt.legend()
plt.tight_layout()
plt.show()
```
